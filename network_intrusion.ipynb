```python
column_names = {col: col.replace(' ', '') for col in final_df.columns}

final_df = final_df.rename(columns=column_names)
```

```python
final_df.columns
```

```python
attack_conditions = [(final_df['flag'] == 1), (final_df['flag'] == 0)]

# Corresponding labels for each condition
attack_labels = ['normal', 'error']

final_df['attack'] = np.select(attack_conditions, attack_labels, default='unknown')

final_df.head()
```

```python
categorical = final_df[['protocol_type', 'service', 'flag']]
```

```python
cont = final_df[list(final_df.columns.difference(['protocol_type', 'service', 'flag'],sort=False))]
```

```python
for i in ['protocol_type', 'service', 'flag']:
    categorical.loc[:,i]=categorical[i].astype('category')
    cols=pd.get_dummies(categorical[i],prefix = i, drop_first = True)
    categorical = pd.concat([categorical, cols], axis = 1)
    categorical.drop(i, axis = 1, inplace = True )
```

```python
categorical.columns
```

```python
categorical.columns=categorical.columns.str.replace('.','_')
```

```python
data_final = pd.concat([cont,categorical],axis=1)
```

```python
data_final
```

```python
feature_cols = data_final.columns.difference(['attack'])
```

```python
feature_cols
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(data_final[feature_cols],data_final['attack'],test_size=0.3, 
                                                    random_state = 123)
```

#### KNN Model

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 5)

knn_classifier.fit(train_X, train_Y)

predictions = knn_classifier.predict(test_X)

accuracy = accuracy_score(test_Y, predictions)
print("Accuracy:", accuracy)
```

#### Gradient Boosting Model

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GB = GradientBoostingClassifier(random_state=42)
GB.fit(train_X, train_Y)
```

```python
GB_train, GB_test = GB.score(train_X , train_Y), GB.score(test_X , test_Y)

print(f"Training Score: {GB_train}")
print(f"Test Score: {GB_test}")
```

#### Naive Bayes Model

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
BNB_model = BernoulliNB()
BNB_model.fit(train_X, train_Y)
```

```python
BNB_train, BNB_test = BNB_model.score(train_X , train_Y), BNB_model.score(test_X , test_Y)

print(f"Training Score: {BNB_train}")
print(f"Test Score: {BNB_test}")
```
